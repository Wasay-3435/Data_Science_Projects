#### practice with real world data

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
import numpy as np

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
print("======zomato sales analysis======")
print("\nshape : ",sales_data.shape)
print("sample data for 1st 3 restaurants:",sales_data[0:3])
```

```python
#total sales per year
print(np.sum(sales_data,axis = 0))
yearly_total = np.sum(sales_data[:, 1:] , axis = 0)
print(yearly_total)
total=np.sum(yearly_total[0:5])
print(total)

#->>> axis = 0("goes down") , ->>> axis = 1("goes across")

#minimum sales
minimum_sales = np.min(sales_data[:,1:], axis = 1 )
print(minimum_sales)
#max sales per year
max_sales = np.max(sales_data[:,1:] , axis = 0)
print(max_sales)
#average sales of restaurant
avg_sales = np.mean(sales_data[:,1:],axis=1)
print(avg_sales)
#cumulativesum
cumsum = np.cumsum(sales_data[:,1:] , axis =1)
print(cumsum)

plt.figure(figsize = (10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("average cumulative sales across all restaurants")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

```python
#vector operations
vector1 = np.array([1,2,3,4])
vector2 = np.array([6,7,8,9])

print("addition of vectors : ",vector1 + vector2)
print("Multiplication of vectors : ",vector1 * vector2)
print("dot product",np.dot(vector1,vector2))

#to convert array to upppercase letters:
restaurant_types = np.array(["biryani","pizza","chicken"])
vectorized_upper = np.vectorize(str.upper)
print("In Upper case : ",vectorized_upper(restaurant_types))

#broadcasting
monthly_avg = sales_data[:,1:] / 12
print(monthly_avg)
```

#### Scatter Plots

```python
# Distribution → Histogram

# Comparison (categories) → Bar Chart

# Proportion of a whole → Pie Chart / Donut Chart

# Trend over time → Line Chart

# Relationship between variables → Scatter Plot


import matplotlib.pyplot as plt
hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [45,50,65,59,63,75,59,80]
plt.scatter(hours_studies,exam_scores,color = 'green',marker = '*' ,label ='Student_data')
plt.xlabel('Hours_studied')
plt.ylabel('exam_scores')
plt.title('relation between exam score and study hours')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#for two groups:
import matplotlib.pyplot as plt
plt.scatter([1,2,3] ,[50,55,60] , color = 'blue' , label = 'class A') #g1
plt.scatter([1,2,3] ,[40,50,52] , color = 'Orange' , label = 'class B') #g2
plt.xlabel('Hours_studied')
plt.ylabel('exam_scores')
plt.title('comparison of two classes')
plt.grid(True)
plt.legend()
plt.show()
```

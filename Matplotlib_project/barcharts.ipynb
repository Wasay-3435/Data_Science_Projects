#### Barcharts ,Piechrats ,Histograms

```python
import matplotlib.pyplot as plt
product = ['A','B','C','S','F']
sales = [700,1300,1250,900,675]
plt.bar(product,sales,color = 'Blue',label = 'sales 2025')
#we can also displays the chart horizontally
#plt.barh(product,sales,color = 'Blue',label = 'sales 2025')
plt.title("Different Category wise sales")
plt.xlabel('product')
plt.ylabel('sales')
plt.legend()
plt.grid(False)
plt.show()
```

#### Piechart

```python
import matplotlib.pyplot as plt
regions = ['North','South','West','East','South_East']
revenue = [3000,2000,1500,1000,600]

plt.pie(revenue , labels = regions,autopct = '%1.1f%%', colors = ['gold','skyblue','lightgreen','coral'])
plt.title('Revenue contribution by regions')
plt.show()
```

#### Histogram

```python
scores = [45,67,56,56,78,66,43,56,78,86,73,69,52,75,81]
plt.hist(scores , bins = 5 ,color = 'purple' , edgecolor = 'black')
plt.xlabel('score range')
plt.ylabel('Number of students')
plt.title('score distribution of students')
plt.show()
```

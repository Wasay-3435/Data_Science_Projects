#### Plotting Graphs:


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #pyplot is one of the module in matplotlib
x = [1,2,3,4,5]
y = [10,15,25,20,30]
plt.plot(x,y)
plt.show()

```

```python
import matplotlib.pyplot as plt
list = ['Monday','Tuesaday','Wednesday','Thursday','Friday']
sell = [4,7,12,23,17]
plt.plot(list,sell)
plt.title("sales for the week")
plt.xlabel('days of the week')
plt.ylabel('sales per day')
plt.show()
```

```python
import matplotlib.pyplot as plt
months = [1,2,3,4,5]
sales = [40,50,25,70,61]
plt.plot(months,sales,color ='green',linestyle ='--',marker ='o',label = '2025 sales data')
plt.title('Monthly sales data report')
plt.legend() # we can also pass location here(it will appear only if we set the label in plot function) and font size of box 
plt.grid()
plt.xlim(1,6)
plt.ylim(20,75)
plt.xlabel('Months')
plt.ylabel('sales Per month')
plt.xticks([1,2,3,4,5] , ('M1','M2','M3','M4','M5'))
plt.show()
```

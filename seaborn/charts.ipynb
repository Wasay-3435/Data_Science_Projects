#### Data Visualization Through Seaborn

##### Scatter_plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#scatter plot => axis_level function
tips = sns.load_dataset('tips')
print(tips.to_string())
sns.scatterplot(data = tips , x= 'total_bill' , y = 'tip',hue = 'sex',style ='time',size = 'size')
#hue is a hidden parameter
```

##### Line_plot

```python
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
temp_df
sns.lineplot(data = temp_df , x = 'year' , y = 'lifeExp') #axis level function 
```

```python
#using rel_plot
sns.relplot(data = temp_df,x='year',y='lifeExp',kind ='line')
```

```python
temp_df = gap[gap['country'].isin(['Brazil','Pakistan','Germany'])]
temp_df
```

```python
sns.relplot(kind='line',data = temp_df,x='year', y ='lifeExp',hue ='country',style ='continent')
```

#### Data Visualization Through Seaborn

##### Scatter_plot

```python
import pandas as pd
import scipy.stats as stats
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#scatter plot => axis_level function
tips = sns.load_dataset('tips')
print(tips.to_string())
sns.scatterplot(data = tips , x= 'total_bill' , y = 'tip',hue = 'sex',style ='time',size = 'size')
#hue is a hidden parameter
```

##### Line_plot

```python
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
temp_df
sns.lineplot(data = temp_df , x = 'year' , y = 'lifeExp') #axis level function 
```

```python
#using rel_plot
sns.relplot(data = temp_df,x='year',y='lifeExp',kind ='line')
```

```python
temp_df = gap[gap['country'].isin(['Brazil','Pakistan','Germany'])]
temp_df
```

```python
sns.relplot(kind='line',data = temp_df,x='year', y ='lifeExp',hue ='country',style ='continent')
```

```python
#facet plot:
sns.relplot(data = tips, x='total_bill',y='tip',col ='sex',kind ='scatter') #col is used to create multiple plots based on unique values in the column,
```

##### Distribution_plot

```python
#hist plot:
#figure level =>  histplot
#axis level => hist,kde,rug plot
sns.histplot(data = tips , x = 'total_bill')
```

```python
#figure level function
sns.displot(data = tips, x ='total_bill',kind ='hist',bins = 10)
```

```python
sns.kdeplot(data = tips, x ='total_bill')
```

```python
sns.displot(data = tips, x ='total_bill',kind ='kde',hue ='sex',fill = True,height = 5,aspect = 2)
```

#### loading titanic data_set

```python
titanic = sns.load_dataset('titanic')
print(titanic.to_string())
```

```python
sns.displot(data = titanic ,x ='age',kind = 'hist',element = 'step',hue = 'survived')
```

```python
#### Rug plot:
sns.kdeplot(data = tips, x ='total_bill')
sns.rugplot(data = tips , x = 'total_bill')
```

```python
#### bivarient kde plot
sns.kdeplot(data = tips,x='tip',y='total_bill')
```

#### Matrix Plots:

```python
# Heatmap
temp_df = gap.pivot(index = 'country',columns = 'year',values = 'lifeExp')
plt.figure(figsize = (15,10))
sns.heatmap(temp_df)

```
